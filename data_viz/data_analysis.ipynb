```python
# Installs and imports
!pip install -Iv matplotlib==3.4.2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from matplotlib import cm
from matplotlib import mlab
import seaborn as sns

# Load in the data
df_top = pd.read_csv('/data/complete_data_top.csv')
df_mid = pd.read_csv('/data/complete_data_mid.csv')
df_bot = pd.read_csv('/data/complete_data_bot.csv')
```

```python
# Constants
WEBSITE_ACCESSOR = 'Website URL'
ITAA_ACCESSOR = ' ITAA'
HAM_ACCESSOR = ' HAM'
LIM_ACCESSOR = ' LIM'
AP_ACCESSOR = ' AP'
BBOX_COLOR = 'orange'
TOP_10K_COLOR = 'xkcd:purple'
MID_10K_COLOR = 'xkcd:green'
BOT_10K_COLOR = 'xkcd:sky blue'
COMBO_COLOR = '#588391'
EMPTY_DATA = 'No Data'

# numpy arrays for data visualization
df_top_itaa = df_top[ITAA_ACCESSOR].to_numpy()
df_top_ham = df_top[HAM_ACCESSOR].to_numpy()
df_top_lim = df_top[LIM_ACCESSOR].to_numpy()
df_top_ap = df_top[AP_ACCESSOR].to_numpy()

df_mid_itaa = df_mid[ITAA_ACCESSOR].to_numpy()
df_mid_ham = df_mid[HAM_ACCESSOR].to_numpy()
df_mid_lim = df_mid[LIM_ACCESSOR].to_numpy()
df_mid_ap = df_mid[AP_ACCESSOR].to_numpy()

df_bot_itaa = df_bot[ITAA_ACCESSOR].to_numpy()
df_bot_ham = df_bot[HAM_ACCESSOR].to_numpy()
df_bot_lim = df_bot[LIM_ACCESSOR].to_numpy()
df_bot_ap = df_bot[AP_ACCESSOR].to_numpy()
```

```python
count_top_ap = 0
sum_top_ap = 0

count_mid_ap = 0
sum_mid_ap = 0

count_bot_ap = 0
sum_bot_ap = 0

for curr_ap in df_top_ap:
  if curr_ap == "-1%":
    continue
  curr_ap = curr_ap.strip('%')
  count_top_ap += 1
  sum_top_ap += float(curr_ap)

for curr_ap in df_mid_ap:
  if curr_ap == "-1%":
    continue
  curr_ap = curr_ap.strip('%')
  count_mid_ap += 1
  sum_mid_ap += float(curr_ap)

for curr_ap in df_bot_ap:
  if curr_ap == "-1%":
    continue
  curr_ap = curr_ap.strip('%')
  count_bot_ap += 1
  sum_bot_ap += float(curr_ap)

barlist = plt.bar(["Top 10k","Middle 10k","Bottom 10k"], 
                  [round(sum_top_ap / count_top_ap, 2), 
                   round(sum_mid_ap / count_mid_ap, 2), 
                   round(sum_bot_ap / count_bot_ap, 2)], width=0.4)
barlist[0].set_color(TOP_10K_COLOR)
barlist[1].set_color(MID_10K_COLOR)
barlist[2].set_color(BOT_10K_COLOR)

plt.text(0, round(sum_top_ap / count_top_ap, 2), 
         round(sum_top_ap / count_top_ap, 2), 
         ha='center', Bbox=dict(facecolor=BBOX_COLOR, alpha =.8))
plt.text(1, round(sum_top_ap / count_top_ap, 2), 
         round(sum_mid_ap / count_mid_ap, 2), 
         ha='center', Bbox=dict(facecolor=BBOX_COLOR, alpha =.8))
plt.text(2, round(sum_top_ap / count_top_ap, 2), 
         round(sum_bot_ap / count_bot_ap, 2), 
         ha='center', Bbox=dict(facecolor=BBOX_COLOR, alpha =.8))
plt.ylim(bottom=0, top=100)
plt.xlabel("Popularity Grouping")
plt.ylabel("Accessibility Percentage")
plt.title("Average Accessibility Percentage Per Popularity Grouping")
```

Figure 1 - Average Accessibility Percentage Per Popularity Grouping (Top, Middle Bottom)

Interestingly, accessbility does not appear to be significantly different between the top 10K websites and the bottom 10K websites, siting at a roughly 62% average across the board (not great)

```python
top_totals_sum = []
top_totals_ap = []
mid_totals_sum = []
mid_totals_ap = []
bot_totals_sum = []
bot_totals_ap = []

# Only consider cases where we had data to get a result for all three metrics
for ind in df_top.index:
  if df_top[ITAA_ACCESSOR][ind] != EMPTY_DATA and df_top[HAM_ACCESSOR][ind] != EMPTY_DATA and df_top[LIM_ACCESSOR][ind] != EMPTY_DATA:
    top_totals_sum.append(float(df_top[ITAA_ACCESSOR][ind]) + float(df_top[HAM_ACCESSOR][ind]) + float(df_top[LIM_ACCESSOR][ind]))
    top_totals_ap.append(float(df_top[AP_ACCESSOR][ind].strip("%")))

for ind in df_mid.index:
  if df_mid[ITAA_ACCESSOR][ind] != EMPTY_DATA and df_mid[HAM_ACCESSOR][ind] != EMPTY_DATA and df_mid[LIM_ACCESSOR][ind] != EMPTY_DATA:
    mid_totals_sum.append(float(df_mid[ITAA_ACCESSOR][ind]) + float(df_mid[HAM_ACCESSOR][ind]) + float(df_mid[LIM_ACCESSOR][ind]))
    mid_totals_ap.append(float(df_mid[AP_ACCESSOR][ind].strip("%")))

for ind in df_bot.index:
  if df_bot[ITAA_ACCESSOR][ind] != EMPTY_DATA and df_bot[HAM_ACCESSOR][ind] != EMPTY_DATA and df_bot[LIM_ACCESSOR][ind] != EMPTY_DATA:
    bot_totals_sum.append(float(df_bot[ITAA_ACCESSOR][ind]) + float(df_bot[HAM_ACCESSOR][ind]) + float(df_bot[LIM_ACCESSOR][ind]))
    bot_totals_ap.append(float(df_bot[AP_ACCESSOR][ind].strip("%")))

fig = plt.figure(figsize=(20,10))
g = sns.jointplot(x=top_totals_sum, y=top_totals_ap, kind='hex', 
                  color=TOP_10K_COLOR, joint_kws=dict(gridsize=15))
g.ax_marg_x.remove()
g.ax_marg_y.remove()
plt.xlabel("Total Metric Score")
plt.ylabel("Accessibility Percentage")
plt.title("Overall Occurrences of Total Metric Score vs AP for Top 10k Websites")
plt.show()

fig = plt.figure(figsize=(20,10))
g = sns.jointplot(x=mid_totals_sum, y=mid_totals_ap, kind='hex',
                  color=MID_10K_COLOR, joint_kws=dict(gridsize=15))
g.ax_marg_x.remove()
g.ax_marg_y.remove()
plt.xlabel("Total Metric Score")
plt.ylabel("Accessibility Percentage")
plt.title("Overall Occurrences of Total Metric Score vs AP for Middle 10k Websites")
plt.show()

fig = plt.figure(figsize=(20,10))
g = sns.jointplot(x=bot_totals_sum, y=bot_totals_ap, kind='hex', 
                  color=BOT_10K_COLOR, joint_kws=dict(gridsize=15))
g.ax_marg_x.remove()
g.ax_marg_y.remove()
plt.xlabel("Total Metric Score")
plt.ylabel("Accessibility Percentage")
plt.title("Overall Occurrences of Total Metric Score vs AP for Bottom 10k Websites")
plt.show()

fig = plt.figure(figsize=(20,10))
totals_sum = top_totals_sum
totals_sum += mid_totals_sum
totals_sum += bot_totals_sum
totals_ap = top_totals_ap
totals_ap += mid_totals_ap
totals_ap += bot_totals_ap
g = sns.jointplot(x=totals_sum, y=totals_ap, kind='hex', color=COMBO_COLOR, 
                  joint_kws=dict(gridsize=15))
g.ax_marg_x.remove()
g.ax_marg_y.remove()
plt.xlabel("Total Metric Score")
plt.ylabel("Accessibility Percentage")
plt.title("Overall Occurrences of Total Metric Score vs AP for all Measured Websites")
plt.show()
```

Figure 5-9: Overall Occurances of Total Metric Score vs AP

What is interesting here is that a linear trend is not observed. Instead, we have a concentration from 1-1.5 metric sum and between 40-60% accessibility. If this trend worked for the whole dataset we could consider the data correlated, but as it stands this is quite evidently not correlated. In other words, the base metrics do not appear to be a good way to determine the accessibility percentage of any given website.

```python
websites = []
aps = []

# Only consider cases where we had data to get a result for all three metrics
for ind in df_top.index:
  if df_top[ITAA_ACCESSOR][ind] != EMPTY_DATA and df_top[HAM_ACCESSOR][ind] != EMPTY_DATA and df_top[LIM_ACCESSOR][ind] != EMPTY_DATA:
    if float(df_top[ITAA_ACCESSOR][ind]) + float(df_top[HAM_ACCESSOR][ind]) + float(df_top[LIM_ACCESSOR][ind]) == 3.0:
      websites.append(f"{df_top[WEBSITE_ACCESSOR][ind]} (top 10k)")
      aps.append(int(df_top[AP_ACCESSOR][ind].strip("%")))

for ind in df_mid.index:
  if df_mid[ITAA_ACCESSOR][ind] != EMPTY_DATA and df_mid[HAM_ACCESSOR][ind] != EMPTY_DATA and df_mid[LIM_ACCESSOR][ind] != EMPTY_DATA:
    if float(df_mid[ITAA_ACCESSOR][ind]) + float(df_mid[HAM_ACCESSOR][ind]) + float(df_mid[LIM_ACCESSOR][ind]) == 3.0:
      websites.append(f"{df_top[WEBSITE_ACCESSOR][ind]} (mid 10k)")
      aps.append(int(df_mid[AP_ACCESSOR][ind].strip("%")))

for ind in df_bot.index:
  if df_bot[ITAA_ACCESSOR][ind] != EMPTY_DATA and df_bot[HAM_ACCESSOR][ind] != EMPTY_DATA and df_bot[LIM_ACCESSOR][ind] != EMPTY_DATA:
    if float(df_bot[ITAA_ACCESSOR][ind]) + float(df_bot[HAM_ACCESSOR][ind]) + float(df_bot[LIM_ACCESSOR][ind]) == 3.0:
      websites.append(f"{df_top[WEBSITE_ACCESSOR][ind]} (bot 10k)")
      aps.append(int(df_bot[AP_ACCESSOR][ind].strip("%")))

barlist = plt.bar(websites, aps, width=0.4)
plt.xticks(rotation = 90)
for i in range(0, 8):
  if i < 4:
    barlist[i].set_color(TOP_10K_COLOR)
  else:
    barlist[i].set_color(MID_10K_COLOR)
  plt.text(i, aps[i], aps[i], ha='center', 
           Bbox=dict(facecolor=BBOX_COLOR, alpha =.8))
plt.ylim(bottom=0, top=100)
plt.xlabel("Website Name")
plt.ylabel("Accessibility Percentage (AP)")
plt.title("Accessibility Percentage for each Website with a 3.0 Combined Metrics Score")
```

Figure 10 - Accessibility Percentage for each Website with a 3.0 Combined Metrics Score

Surprisingly, not a single website from the bottom 10k managed to achieve a combined metric score of 3.0. Even more interestingly, the average accessibility percentage is higher for websites from the middle 10k compared to websites in the top 10k. Again, the fact that all of these are not 100% should make it demonstrably clear that these metrics are likely not correlated with the accessibility percentage of a website.

```python
websites_total_above_50 = 0
websites_total_below_50 = 0
for ind in df_top.index:
  if int(df_top[AP_ACCESSOR][ind].strip("%")) > 50:
    websites_total_above_50 += 1
  else:
    websites_total_below_50 += 1

for ind in df_mid.index:
  if int(df_mid[AP_ACCESSOR][ind].strip("%")) > 50:
    websites_total_above_50 += 1
  else:
    websites_total_below_50 += 1
    
for ind in df_bot.index:
  if int(df_bot[AP_ACCESSOR][ind].strip("%")) > 50:
    websites_total_above_50 += 1
  else:
    websites_total_below_50 += 1

plt.pie([websites_total_above_50, websites_total_below_50], 
        labels=[">50% AP", "<=50% AP"], colors=[TOP_10K_COLOR, MID_10K_COLOR],
        explode=[0.2, 0], shadow=True, autopct='%1.1f%%')
plt.title("Proportion of Website with an Accessibility Percentage of > 50%")
```
